### Aggregating the rows by the country

### Calculating the first derivative of the curve

```python
corona_dataset_aggregated.loc['China'].diff().plot()
```

### Find maximum infection rate for China

```python
corona_dataset_aggregated.loc['China'].diff().max()
```

```python
corona_dataset_aggregated.loc['India'].diff().max()
```

### Find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())

corona_dataset_aggregated['Max infection rate'] = max_infection_rates
```

```python
corona_dataset_aggregated.head(10)
```

### Creating a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggregated['Max infection rate'])
```

```python
corona_data.head(10)
```



# Next Task: Correlation of Happiness Report Against Death Reports

### IMPORTING DEATH REPORT DATASET 

```python
deaths_report = pd.read_csv('covid19_deaths_dataset.csv')
```

```python
deaths_report.head()
```

### Dropping useless columns

```python
deaths_report.drop(['Lat','Long'],axis = 1, inplace = True)
```

### Aggregation

```python
deaths_aggregated = deaths_report.groupby('Country/Region').sum()
```

```python
deaths_aggregated.head(15)
```

### Plotting Number of Death in China

```python
deaths_aggregated.loc['China'].plot()
```

 ### Finding Max death rate in a single day in China

```python
deaths_aggregated.loc['China'].diff().max()
```

### Adding a column with Max Rate of Death in a day

```python
countries = list(deaths_aggregated.index)
max_deaths = []

for c in countries:
    max_deaths.append(deaths_aggregated.loc[c].diff().max())

max_deaths
deaths_aggregated['Max Death Rate'] = max_deaths
```

```python
deaths_aggregated.head(20)
```

### Finalizing Death Dataset with needed column

```python
death_report = pd.DataFrame(deaths_aggregated['Max Death Rate'])
```

```python
death_report.head(20)
```

### Merging Datasets

```python
ds = death_report.join(happiness_report, how = 'inner')
ds
```


# Plotting Death Rate against factors of Happiness
